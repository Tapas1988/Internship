```python
# Cheking Data distribution in each column
plt.figure(figsize=(10,12),facecolor='y')
plotnumber=1
for column in dt:
    if plotnumber<=36:
        ax=plt.subplot(9,4,plotnumber)
        sns.distplot(dt[column])
        plt.xlabel(column)
        
    plotnumber+=1
plt.show()
```

```python
# Observation
# Each columns have outlier. Data is not normal distributed. there are right skewness in each column.
```

```python
# find outlier in each & every cloumn using box plot
# Cheking Data distribution in each column
plt.figure(figsize=(10,12),facecolor='y')
plotnumber=1
for column in dt:
    if plotnumber<=36:
        ax=plt.subplot(9,4,plotnumber)
        sns.boxplot(dt[column])
        plt.xlabel(column)
        
    plotnumber+=1
plt.show()
```

```python
# Observation
# Outlier present in every column
```

```python
dt.columns
```

```python
# minimize the outlier using by z distribution
from scipy.stats import zscore
z_score=zscore(dt['Meningitis'])
abs_z_score=np.abs(z_score)
filter=(abs_z_score<3)
dt=dt[filter]
```

```python
# minimize the outlier using by z distribution
from scipy.stats import zscore
z_score=zscore(dt[["Alzheimer's Disease and Other Dementias","Parkinson's Disease",'Nutritional Deficiencies','Malaria','Drowning','Interpersonal Violence','Maternal Disorders','HIV/AIDS']])
abs_z_score=np.abs(z_score)
filter=(abs_z_score<3).all(axis=1)
dt=dt[filter]
```

```python
# minimize the outlier using by z distribution
from scipy.stats import zscore
z_score=zscore(dt[['Drug Use Disorders', 'Tuberculosis', 'Cardiovascular Diseases',
       'Lower Respiratory Infections', 'Neonatal Disorders',
       'Alcohol Use Disorders', 'Self-harm', 'Exposure to Forces of Nature',
       'Diarrheal Diseases', 'Environmental Heat and Cold Exposure',
       'Neoplasms', 'Conflict and Terrorism', 'Diabetes Mellitus',
       'Chronic Kidney Disease', 'Poisonings', 'Protein-Energy Malnutrition',
       'Road Injuries', 'Chronic Respiratory Diseases',
       'Cirrhosis and Other Chronic Liver Diseases', 'Digestive Diseases',
       'Fire, Heat, and Hot Substances', 'Acute Hepatitis']])
abs_z_score=np.abs(z_score)
filter=(abs_z_score<3).all(axis=1)
dt=dt[filter]
```

```python
# find outlier in each & every cloumn using box plot
# Cheking Data distribution in each column
plt.figure(figsize=(10,12),facecolor='y')
plotnumber=1
for column in dt:
    if plotnumber<=36:
        ax=plt.subplot(9,4,plotnumber)
        sns.boxplot(dt[column])
        plt.xlabel(column)
        
    plotnumber+=1
plt.show()
```
