```python
# Relationship checking in feature- one featur to othere feature
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x,i) for i in range(x.shape[1])]
vif['Feature']=x.columns
vif
```

```python
from sklearn.feature_selection import SelectPercentile,chi2
```

```python
percentile=SelectPercentile(score_func=chi2,percentile=70)
percentile=percentile.fit(x,y)
```

```python
cols=percentile.get_support(indices=True)
print('feature=',cols)
```

```python
feature=x.columns[cols]
print('feature=',list(feature))
```
