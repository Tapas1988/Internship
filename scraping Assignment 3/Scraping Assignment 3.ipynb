```python
brand=[]    #brand name extract of Guitar
start=0
end=3
for page in range(start,end):
    brand_name=driver.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]')
    for i in brand_name:
        brand.append(i.text)
    next=driver.find_element(By.XPATH,'//span[@class="s-pagination-item s-pagination-selected"]')
    next.click()
    

brand
    
```

```python
len(brand)
```

```python
product_name=[]    #product name extract of Guitar
start=0
end=3
for page in range(start,end):
    product=driver.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]')
    for i in product[1:]:
        product_name.append(i.text)
    next=driver.find_element(By.XPATH,'//span[@class="s-pagination-item s-pagination-selected"]')
    next.click()
product_name
```

```python
len(product_name)
```

```python
product_price=[]    #product price extract of Guitar
start=0
end=3
for page in range(start,end):
    price=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    for i in price:
        product_price.append(i.text)
    next=driver.find_element(By.XPATH,'//span[@class="s-pagination-item s-pagination-selected"]')
    next.click()
product_price
```

```python
len(product_price)
```

```python
delivery=[]    #product delivery data extract of Guitar
start=0
end=3
for page in range(start,end):
    expt_delivery=driver.find_elements(By.XPATH,'//span[@class="a-color-base a-text-bold"]')
    for i in expt_delivery:
        delivery.append(i.text)
    next=driver.find_element(By.XPATH,'//span[@class="s-pagination-item s-pagination-selected"]')
    next.click()
delivery
```

```python
len(delivery)
```

```python
url=[]    #product delivery data extract of Guitar
start=0
end=3
for page in range(start,end):
    expt_url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in expt_url:
        url.append(i.get_attribute('href'))
    next=driver.find_element(By.XPATH,'//span[@class="s-pagination-item s-pagination-selected"]')
    next.click()
url
```
