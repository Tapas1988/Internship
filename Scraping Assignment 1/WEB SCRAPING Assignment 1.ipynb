```python
url_imdb_top_100=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
url_imdb_top_100
                              
```

```python
top_100movies=BeautifulSoup(url_imdb_top_100.content)
top_100movies
```

```python
first_top_movies=top_100movies.find('h3',class_="lister-item-header")
first_top_movies
```

```python
#         2. a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#            and make data frame.
######################################################################################################

#  100 Movies Name:-
###################################################
```

```python
top_100_movies=[]
for i in top_100movies.find_all('h3', class_="lister-item-header",limit=100):
    top_100_movies.append(i.text)
top_100_movies
```

```python
#         2. a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#            and make data frame.
######################################################################################################

# Rateing of IMDB Movies
```

```python
First_Rateing=top_100movies.find('div',class_="inline-block ratings-imdb-rating")
```

```python
First_Rateing
```

```python
IMDB_Rating=[]
for i in top_100movies.find_all('div',class_="inline-block ratings-imdb-rating",limit=100):
    IMDB_Rating.append(i.text)
    
IMDB_Rating
```

```python
#         2. a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#            and make data frame.
######################################################################################################

# Years of Release

```

```python
year_relse=[]
for i in top_100movies.find_all('span', class_="lister-item-year text-muted unbold",limit=100):
    year_relse.append(i.text)
year_relse
```

```python
#         2. a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#            and make data frame.
######################################################################################################

# DataFrame of Top IMDB movies
```

```python
import pandas as pd
```

```python
df_movies=pd.DataFrame({'Movies Name':top_100_movies,
                 'Reating':IMDB_Rating,
                 'Year of release':year_relse})
df_movies
```

```python
#        3  a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
#            release) and make data frame.
######################################################################################################


################# for getting all name of Indian top movies  ##################

```

```python
#        3  a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
#            release) and make data frame.
######################################################################################################

######     for getting all rating of Indian top movies  ############

```

```python
Indian_top_movies_reating=[]
for i in soup_Indian_movies.find_all('td', class_="ratingColumn imdbRating",limit=100):
    Indian_top_movies_reating.append(i.text)
    
Indian_top_movies_reating
```

```python
#        3.  a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
#            release) and make data frame.
######################################################################################################


# ######      for getting all Release of the Year of Indian top movies   #########

```

```python
Indian_Movies_Release_year=[]
for i in soup_Indian_movies.find_all('span', class_="secondaryInfo",limit=100):
    Indian_Movies_Release_year.append(i.text)
    
Indian_Movies_Release_year
```

```python
#        3  a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
#            release) and make data frame.
######################################################################################################

##########   Create Data Frame ################

```

```python
# ####   4. display list of respected former presidents of India(i.e. Name , Term of office)

######################################################################################################
```

```python
former_presidents=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
former_presidents
```

```python
# ####   4. display list of respected former presidents of India(i.e. Name , Term of office)

######################################################################################################

#                     Created Name of Presidents
#######################################################################################################
```

```python
president_name=[]
for i in soup_presidents.find_all('div',class_="presidentListing"):
    president_name.append(i.text.splitlines(keepends=False)[1:2])
    
president_name
```

```python
# ####   4. display list of respected former presidents of India(i.e. Name , Term of office)

######################################################################################################

###########      Term of office  ###################
```

```python
president_office=[]
for i in soup_presidents.find_all('span',class_="terms"):
    president_office.append(i.text)
president_office
```

```python
# ####   4. display list of respected former presidents of India(i.e. Name , Term of office)

######################################################################################################

#########             DataFrame Former presidents of India     ###########
```

```python
former_president_IND=pd.DataFrame({'President_name':president_name,
                              'term of Office':president_office})
former_president_IND
```

```python


#    10.  python program to scrape the details of top publications from Google Scholar from
#         https://scholar.google.com/citations?view_op=top_venues&hl=en
############################################################################################

```

```python
url_publication=requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
url_publication
```

```python
soup_publication=BeautifulSoup(url_publication.content)
soup_publication
```

```python
#    10.  python program to scrape the details of top publications from Google Scholar from
#         https://scholar.google.com/citations?view_op=top_venues&hl=en
    ######################################################################

#                  i) Rank

################################################################
```

```python
publication_rank=[]
for i in soup_publication.find_all('td', class_="gsc_mvt_p"):
    publication_rank.append(i.text)
    
publication_rank
```

```python
#    10.  python program to scrape the details of top publications from Google Scholar from
#         https://scholar.google.com/citations?view_op=top_venues&hl=en
    ######################################################################


#                          ii) Publication

###########################################################
```

```python
publication=[]
for i in soup_publication.find_all('td', class_="gsc_mvt_t"):
    publication.append(i.text)
    
publication
```

```python
#    10.  python program to scrape the details of top publications from Google Scholar from
#         https://scholar.google.com/citations?view_op=top_venues&hl=en
    ######################################################################

#                    iii) h5-index

#####################################################################
```

```python
publication_h5_index=[]
for i in soup_publication.find_all('a', class_="gs_ibl gsc_mp_anchor"):
    publication_h5_index.append(i.text)
    
publication_h5_index
```

```python
#    10.  python program to scrape the details of top publications from Google Scholar from
#         https://scholar.google.com/citations?view_op=top_venues&hl=en
    ######################################################################                   
    
    
                     # iv) h5-median
##############################################################
```

```python
publication_h5_median=[]
for i in soup_publication.find_all('span', class_="gs_ibl gsc_mp_anchor"):
    publication_h5_median.append(i.text)
    
publication_h5_median
```

```python
#                        Creat DataFrame
##########################################################
```

```python
df_Publication=pd.DataFrame({'Rank':publication_rank,
                            'Publication':publication,
                            'h5_index':publication_h5_index,
                            'h5_median':publication_h5_median})

df_Publication
```
